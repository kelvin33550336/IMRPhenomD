```python
# Load catalog from file

with open('catalog.json') as file:
    metadata_list = json.load(file)
metadata_df = pd.DataFrame(metadata_list)

display(metadata_df)
```

```python
# Filter data of precessing BBH events and output readable results

data = pd.DataFrame([])
# catalog_droplist = ['SXS:BBH:0370', 'SXS:BBH:2092', 'SXS:BBH:0330', 'SXS:BBH:2116', 'SXS:BBH:2111', 'SXS:BBH:0335', 'SXS:BBH:2114', 'SXS:BBH:0263', 'SXS:BBH:2133', 'SXS:BBH:1932', 'SXS:BBH:0304', 'SXS:BBH:0327', 'SXS:BBH:2123', 'SXS:BBH:2128', 'SXS:BBH:2132', 'SXS:BBH:2153', 'SXS:BBH:0045', 'SXS:BBH:0292', 'SXS:BBH:0156', 'SXS:BBH:0151', 'SXS:BBH:0001', 'SXS:BBH:1418', 'SXS:BBH:0167', 'SXS:BBH:1419', 'SXS:BBH:0063', 'SXS:BBH:0172', 'SXS:BBH:0152', 'SXS:BBH:0001', 'SXS:BBH:1417', 'SXS:BBH:0167', 'SXS:BBH:1426', 'SXS:BBH:0063']

for i in range(metadata_df.shape[0]):
    row = metadata_df.iloc[i]
    try: 
        cond = (np.abs(row['reference_dimensionless_spin1'][0]) < 5e-3) & \
               (np.abs(row['reference_dimensionless_spin1'][1]) < 5e-3) & \
               (np.abs(row['reference_dimensionless_spin2'][0]) < 5e-3) & \
               (np.abs(row['reference_dimensionless_spin2'][1]) < 5e-3) & \
               (row['object_types'] == 'BHBH') 

        eccentricity = row['reference_eccentricity']
        if cond & (eccentricity < 2e-3):
            data = pd.concat([data, row], axis=1)

    except Exception:
        pass

    try: 
        cond = (np.abs(row['reference_dimensionless_spin1'][0]) < 5e-3) & \
               (np.abs(row['reference_dimensionless_spin1'][1]) < 5e-3) & \
               (np.abs(row['reference_dimensionless_spin2'][0]) < 5e-3) & \
               (np.abs(row['reference_dimensionless_spin2'][1]) < 5e-3) & \
               (row['object_types'] == 'BHBH') 

        eccentricity = row['reference_eccentricity']
        if cond & (eccentricity[0] == '<') & (float(eccentricity[1:]) < 2e-3):
            row.loc['reference_eccentricity'] = eccentricity[1:]
            data = pd.concat([data, row], axis=1)

    except Exception:
        pass
            
data = data.T[['name', 'reference_mass_ratio', 'reference_dimensionless_spin1', 'reference_dimensionless_spin2', 'reference_chi_eff', 'reference_eccentricity', 'number_of_orbits']].sort_values(by=['reference_mass_ratio'])
display(data)
```

```python
# Change names to be 4 digit numbers, and only get chi_z

for i in range(data.shape[0]):
    sxs_name = []
    sxs_name.extend(data.iloc[i]['name'])
    data.iloc[i]['name'] = ''.join(sxs_name[-4:])
    
    chi1 = data.iloc[i]['reference_dimensionless_spin1'][2]
    data.iloc[i]['reference_dimensionless_spin1'] = chi1
    chi2 = data.iloc[i]['reference_dimensionless_spin2'][2]
    data.iloc[i]['reference_dimensionless_spin2'] = chi2
```

```python
filtered_data1 = pd.DataFrame([])
filtered_data2 = pd.DataFrame([])
filtered_data3 = pd.DataFrame([])
filtered_data4 = pd.DataFrame([])
for i in range(data.shape[0]):
    row = data.iloc[i]
    cond1 = (row['reference_dimensionless_spin1'] > 0.0) & (row['reference_dimensionless_spin2'] > 0.0) # 1pos2pos
    cond2 = (row['reference_dimensionless_spin1'] < 0.0) & (row['reference_dimensionless_spin2'] > 0.0) # 1neg2pos
    cond3 = (row['reference_dimensionless_spin1'] < 0.0) & (row['reference_dimensionless_spin2'] < 0.0) # 1neg2neg
    cond4 = (row['reference_dimensionless_spin1'] > 0.0) & (row['reference_dimensionless_spin2'] < 0.0) # 1pos2neg

    try: 
        if cond1:
            filtered_data1 = pd.concat([filtered_data1, row], axis=1)
        elif cond2:
            filtered_data2 = pd.concat([filtered_data2, row], axis=1)
        elif cond3:
            filtered_data3 = pd.concat([filtered_data3, row], axis=1)
        elif cond4:
            filtered_data4 = pd.concat([filtered_data4, row], axis=1)
    except TypeError:
        pass

filtered_data1 = filtered_data1.T.sort_values(by=['reference_mass_ratio'])
filtered_data2 = filtered_data2.T.sort_values(by=['reference_mass_ratio'])
filtered_data3 = filtered_data3.T.sort_values(by=['reference_mass_ratio'])
filtered_data4 = filtered_data4.T.sort_values(by=['reference_mass_ratio'])

display(filtered_data1)
```
